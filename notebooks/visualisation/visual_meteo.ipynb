```python
meteo_sorted= pd.read_csv('./data/prepared/meteo_clean.csv')
```

##### Variation de la température au fil du temps

```python
plt.figure(figsize=(12, 6))
plt.plot(meteo_sorted['date'], meteo_sorted['temperature'], label='Température (K)', color='orange')
plt.xlabel('Date')
plt.ylabel('Température (K)')
plt.title('Variation de la température au fil du temps')
plt.legend()
plt.grid()
plt.show()
```

###### 

#### Distribution de l'humidité

```python
plt.figure(figsize=(8, 6))
sns.histplot(meteo_sorted['humidity'], bins=30, kde=True, color='blue')
plt.xlabel('Humidité (%)')
plt.ylabel('Fréquence')
plt.title('Distribution de l\'humidité')
plt.show()
```

#### Direction moyenne du vent

```python
plt.figure(figsize=(10, 6))
plt.hist(meteo_sorted['wind_direction'], bins=36, color='green', edgecolor='black')
plt.xlabel('Direction du vent (degrés)')
plt.ylabel('Fréquence')
plt.title('Distribution de la direction du vent')
plt.grid()
plt.show()

```

#### Pression au niveau de la mer au fil du temps

```python
plt.figure(figsize=(12, 6))
plt.plot(meteo_sorted['date'], meteo_sorted['pressure_sea_level'], label='Pression au niveau de la mer (Pa)', color='purple')
plt.xlabel('Date')
plt.ylabel('Pression (Pa)')
plt.title('Variation de la pression au niveau de la mer')
plt.legend()
plt.grid()
plt.show()
```

#### Comparaison entre la vitesse du vent et la température

```python
plt.figure(figsize=(10, 6))
plt.scatter(meteo_sorted['wind_speed'], meteo_sorted['temperature'], alpha=0.6, color='red')
plt.xlabel('Vitesse du vent (m/s)')
plt.ylabel('Température (K)')
plt.title('Relation entre la vitesse du vent et la température')
plt.grid()
plt.show()
```
